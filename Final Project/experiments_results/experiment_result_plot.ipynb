```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
#experiment result - which method train fast and good
discrete_prompt = {'acc':[0.0753], 'exe_time':[0]}
deep_vision_prompt = {'acc':[0.409, 0.78, 0.82, 0.864, 0.879, 0.864, 0.894], 'exe_time':[30, 47, 66, 83, 155, 255, 433]}
language_prompt = {'acc':[0.303, 0.72, 0.78, 0.795], 'exe_time':[59, 187, 354, 875]}


#maple = {'acc':[0.235, 0.856, 0.848, 0.492], 'exe_time':[39, 109, 122, 480]}
maple = {'acc':[0.545, 0.705, 0.856, 0.848, 0.841], 'exe_time':[29, 80, 109, 122, 310]}


#resnet = {'acc':[0.0758, 0.0758, 0.0758, 0.0758, 0.0758], 'exe_time':[76, 137, 200, 263, 324]}
resnet = {'acc':[0.0758, 0.0909, 0.0985, 0.1136, 0.1061], 'exe_time':[76, 137, 200, 263, 324]}


vit = {'acc':[0.5758, 0.7197, 0.8030, 0.8258, 0.8333], 'exe_time':[75, 137, 196, 259, 318]}

data = [discrete_prompt, deep_vision_prompt, language_prompt, maple, resnet, vit]
```

```python
for exp in data:
    for i in range(len(exp['acc'])):
        exp['acc'][i] = exp['acc'][i] * 100
```

```python

graph = sns.lineplot(deep_vision_prompt, x='exe_time', y='acc',err_style=None)
sns.lineplot(language_prompt, x='exe_time', y='acc',err_style=None)
sns.lineplot(maple, x='exe_time', y='acc',err_style=None)
graph.axhline(7.53, linestyle='--', color='grey')
sns.lineplot(resnet, x='exe_time', y='acc',err_style=None)
sns.lineplot(vit, x='exe_time', y='acc',err_style=None)


plt.legend(labels=['Vision Prompt', 'Language Prompt', 'Vision-Langauge Prompt', 'Discrete Prompt', 'Finetune ResNet50', 'Finetune ViT'])
y_labels = graph.get_yticks()
yticks = ticker.FormatStrFormatter('%.0f%%')
graph.yaxis.set_major_formatter(yticks)
plt.ylabel('Accuracy')
plt.xlabel('Training time (sec)')
```
